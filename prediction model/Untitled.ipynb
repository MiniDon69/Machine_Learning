```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
import pandas as pd

dataset = pd.read_csv("predictive_maintenance.csv")
print("dataset imported")
# Preprocess the categorical 'Type' column
dataset['Type'] = dataset['Type'].map({'H': 0, 'L': 1, 'M': 2})
print("type['H','L','M'] converted to type[0,1,2]")
# Separate features (X) and target (y)
X = dataset[['Type', 'Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']]
y = dataset['Failure Type']

# Convert categorical 'Failure Type' to numerical labels
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
print("converting categorical ouput to numeric")

# Build the model
model = Sequential()
model.add(Input(shape=(X.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(len(label_encoder.classes_), activation='softmax'))  # Output layer with softmax for multi-class classification

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=50, batch_size=32, validation_split=0.2)

# Export the model
model.save('failure_prediction_model.h5')
print("Model saved")
```

```python

from tensorflow.keras.models import load_model
from sklearn.metrics import accuracy_score, precision_score
from tensorflow.keras.models import Sequential
# Load the saved model
model = load_model('failure_prediction_model.h5')

# Assuming X_test and y_test are your test data
# You need to preprocess X_test similar to how you preprocessed the training data
# (e.g., map 'Type' to numerical values)

# For demonstration, let's assume you have a DataFrame named 'test_dataset'
# with the same columns as the training data
test_dataset = pd.read_csv('predictive_maintenance.csv')  # Replace with your actual test data

# Preprocess 'Type' column
test_dataset['Type'] = test_dataset['Type'].map({'H': 0, 'L': 1, 'M': 2})

X_test = test_dataset[['Type', 'Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']]
y_test = label_encoder.transform(test_dataset['Failure Type'])

# Make predictions
y_pred = model.predict_classes(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Evaluate precision
precision = precision_score(y_test, y_pred, average='weighted')
print(f"Precision: {precision:.2f}")
```
